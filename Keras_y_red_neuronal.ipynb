```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
X.shape, y.shape
```

```python
from keras.layers import Dense, Activation, Input
from keras.models import Sequential, Model
```

##Forma 1

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

##Forma 2

```python
model = Sequential()
model.add(Dense(12, input_dim=8))
model.add(Activation('relu'))
model.add(Dense(6))
model.add(Activation('relu'))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
# Otra forma de hacer, combinando la forma 1 y la forma 2
model = Sequential()
model.add(Dense(12, input_dim=8))
model.add(Activation('relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

##Forma 3


```python
model = Sequential([
    Dense(12, input_dim=8, activation='relu'),
    Dense(6, activation='relu'),
    Dense(1, activation='sigmoid')
    ])
```

##Forma 4

```python
model = Sequential([
    Dense(12, input_dim=8),
    Activation('relu'),
    Dense(6),
    Activation('relu'),
    Dense(1),
    Activation('sigmoid')
])
```

##Forma 5

```python
I = Input(shape=(8,))
L = Dense(12, activation='relu')(I)
L = Dense(6, activation='relu')(L)
salida = Dense(1, activation='sigmoid')(L)
model = Model(inputs=I, outputs=salida)
```

##Forma 6

```python
I = Input(shape=(8,))
L = Dense(12)(I)
L = Activation('relu')(L)
L = Dense(6)(L)
L = Activation('relu')(L)
salida = Dense(1)(L)
salida = Activation('sigmoid')(salida)
model = Model(inputs=I, outputs=salida)
```

##Forma 7

```python
I = Input(shape=(8,))
densa12 = Dense(12, activation="relu")
densa6 = Dense(6,activation="relu")
salida = Dense(1, activation="sigmoid")

capas = [I, densa12, densa6, salida]
model = Sequential(capas)
```

##Forma 8

```python
I = Input(shape=(8,))
densa12 = Dense(12, activation="relu")
densa6 = Dense(6,activation="relu")
salida = Dense(1, activation="sigmoid")

capas = salida(densa6(densa12(I)))
model = Model(inputs=I, outputs=capas)
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=150)
```

```python
score = model.evaluate(X, y)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
