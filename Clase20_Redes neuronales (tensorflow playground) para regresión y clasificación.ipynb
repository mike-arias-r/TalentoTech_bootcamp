GRADIENTE DECENDIENTE

```python
import numpy as np
import matplotlib.pyplot as plt

#define la funcion a optimizar
funcion = lambda x: x**2 - 3*x + 2

# crear un rango de valores x para graficar
x= np.linspace(-10,10,200)

#graficar la funcion
plt.plot(x,funcion(x))
plt.title("Grafico de la funcion X2 + 3x + 2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()

#definir el gradiente de la funcion
#gradiente = lambda x: 2*x - 3
```

```python
# Defina la derivada de la funcion

def derivada(x):
    return 2*x - 3

# Defina la funcion de descenso del gradiente
def grad(x_inicio, precision, l_r):
    x_list, y_list = [x_inicio], [funcion(x_inicio)]
    while True:
      d_x = derivada(x_inicio)
      xi = x_inicio
      x_inicio = x_inicio - (l_r * d_x)
      print(f"xi = {xi}, xi+1 = {x_inicio}")
      x_list.append(x_inicio)
      y_list.append(funcion(x_inicio))
      if abs(x_list[-1] - x_list[-2]) <= precision:
        break
    print("precision de : " + str(precision))
    print("Tasa de aprendizaje de : " + str(l_r))
    print("El minimo local ocurre en x: "+ str(x_inicio) + ", y:" + str(funcion(x_inicio)))
    print("numero de pasos: " + str(len(x_list)))
    plt.scatter(x, funcion(x))
    plt.scatter(x_list, y_list, c="r")
    plt.plot(x_list, y_list, c="r")
    plt.plot(x, funcion(x))
    plt.title("Grafico del descenso del gradiente paso a paso")
    plt.show()

#ejecuta el descenso del gradiente con diferentes valores iniciales, precision y tasa de aprendizaje
grad(x_inicio=-15, precision=0.00000005, l_r=0.1)
#grad(x_inicio=15, precision=0.0001, l_r=0.01)
```
